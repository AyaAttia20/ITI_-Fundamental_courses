## Question 3

```python
data_telco=pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
data_telco.head()
```

##### set index of df CustomerID

```python
data_telco.set_index('customerID', inplace=True)
```

```python
data_telco.head()
```

```python
data_telco.info()
```

```python
data_telco.isna().sum()
```

#### convert type of "TotalCharges" column from object to numeric

```python
data_telco['TotalCharges']= pd.to_numeric(data_telco['TotalCharges'],errors='coerce')
```

```python
data_telco['TotalCharges'].dtype
```

### fill missing values  in TotalCharges

```python
data_telco['TotalCharges']=data_telco['TotalCharges'].fillna(value=data_telco['TotalCharges'].mean())
```

#### encoding the categorical columns expect ['Contract','Churn']

```python
encod_df=pd.get_dummies(data_telco.drop(['Contract','Churn'],axis=1))
```

```python
encod_df.head()
```

## analyze "Contract" column

#### check what are unique values in contract column

```python
data_telco['Contract'].unique()
```

#### check what are values  counts in each unique in  contract colum

```python
data_telco['Contract'].value_counts()
```

### ordinal encoding on Contract column

```python
ordinal_contract = {"Month-to-month":1, "One year":2, "Two year":3}
encod_df['Contract'] = data_telco['Contract'].replace(ordinal_contract)
ordinal_contract
```

```python
encod_df['Contract']
```

```python
encod_df.head()
```

### ordinal encoding on Churn column

```python
ordinal_contract_churn = {"No":0, "Yes":1}
encod_df['Churn']= data_telco['Churn'].replace(ordinal_contract_churn)
ordinal_contract_churn
```

```python
encod_df['Churn']
```

```python
encod_df['Churn'].dtype
```

```python
encod_df['Churn'].isna().sum()
```

### Now ,check if there any missing values in all columns in df

```python
encod_df.isna().sum()
```

#### create dataframe contains the Churn values as target

```python
target=encod_df['Churn']
target
```

```python
target_df=pd.DataFrame(target)
```

```python
target_df
```

##### create dataframe contains all columns without the Churn

```python
data=encod_df.drop('Churn',axis=1)
data
```

##### Normalization

##### check all data type is numeric

```python
encod_df.info()
```

#### normalization for all numerical columns

```python
encod_df_norm=(encod_df-encod_df.min())/(encod_df.max()-encod_df.min())
encod_df_norm.head()
```

### save the target dataframe in csv

```python
target_df.to_csv('target_df.csv')
```

### save the normalized dataframe in csv

```python
encod_df_norm.to_csv('encod_df_norm.csv')
```
